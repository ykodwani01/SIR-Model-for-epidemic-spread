```python
import numpy as np
import matplotlib.pyplot as mp
import math as m
```

```python
def euler_integrate_1(f, a, b,  init, del_t,ta):
    n = int((b-a)/del_t)
    x = np.linspace(a, b, n+1)
    y = np.zeros(n+1)

    y[0] = init
    for i in range(1, n+1):
        y[i] = y[i-1] + del_t * f(y[i-1],ta)
    return y

def euler_integrate_2(a, b,x0,y0, del_t):

    n = int((b-a)/del_t) -1
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    x[0] = x0
    y[0]= y0

    for i in range(1, n+1):
        x[i] = x[i-1] + del_t*(A*x[i-1]*y[i-1] - B*x[i-1])
        y[i] = y[i-1] - del_t*A*x[i-1]*y[i-1]
    return y,x

def euler_integrate_3(a, b,x0,y0,z0, del_t):

    n = int((b-a)/del_t) -1
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    z = np.zeros(n+1)
    x[0] = x0
    y[0]= y0
    z[0]=z0

    for i in range(1, n+1):
        x[i] = x[i-1] + del_t*(A*x[i-1]*y[i-1] - B*x[i-1])
        y[i] = y[i-1] - del_t*A*x[i-1]*y[i-1]
        z[i] = z[i-1] + del_t*B*x[i-1]

    return y,x,z

```

```python
N = 100000
x0 = 10
y0 = N-x0
z0 = 0
A = 0.0001
B=0.01

init=0
# initValue=0

del_t = 0.005 #delta t is in day

y_euler,x_euler = euler_integrate_2(init, 1000*del_t,x0,y0, del_t)


# timeRange = np.arange(init,59*del_t,del_t)
mp.plot(y_euler,x_euler,linewidth=1.5)

mp.grid(which='major', color='#DDDDDD', linewidth = 1)
mp.xlabel('y(t)')
mp.ylabel('x(t)')
# mp.title(x(t))

mp.show()

```

```python
init=0
# initValue=0
A=0.00001
B = 0.1
del_t = 0.5 #delta t is in day

y_euler,x_euler,z_euler = euler_integrate_3(init, 100*del_t,x0,y0,z0, del_t)

timeRange = np.arange(init,100*del_t,del_t)
mp.plot(timeRange,x_euler,linewidth=1.5)
mp.plot(timeRange,y_euler,linewidth=1.5)
mp.plot(timeRange,z_euler,linewidth=1.5)
mp.xlabel('t')
mp.ylabel('x(t), y(t)  and z(t)')
mp.legend(["x(t)", "y(t)" ,"z(t)"])
mp.grid()
mp.show()
```
